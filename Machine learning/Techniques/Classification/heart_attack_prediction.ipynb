```python
import pandas as pd 
```

```python
df = pd.read_csv("heart_attack_prediction_dataset.csv")
df
```

```python
print("The Shape of the dataset: " , df.shape)
```

```python
print("Summary Statistics: ",df.describe())
```

```python
print("info about columns: ",df.info())
```

```python
print("Missing values in each columns: ",df.isnull().sum())
```

```python
categorical_cols= df.select_dtypes(include ="object").columns
print("categorical columns : ",categorical_cols)
```

```python
df = df.drop('Patient ID',axis=1).copy()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Sex'] = le.fit_transform(df['Sex'])
```

```python
multi_cat = ['Blood Pressure', 'Diet', 'Country', 'Continent',
       'Hemisphere']
df = pd.get_dummies(df,columns = multi_cat,drop_first =True)
```

```python
print("Encoded columns: ",df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
numerical_cols = ['Age', 'Cholesterol', 'Heart Rate', 
                  'Exercise Hours Per Week','BMI']
```

```python
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
print(df.head())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop("Heart Attack Risk",axis=1)
y = df["Heart Attack Risk"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.10,random_state=42)
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("Training State: ",X_train.shape)
print("Test Set: ",X_test.shape)
```

```python
print("y_train shape: ",y_train.shape)
print("y_test shape: ",y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'C' : [0.01,0.1,1,10],
              'penalty' : ['l2'],
              'solver' : ['lbfgs','liblinear']}
```

```python
logic = LogisticRegression(max_iter =1000,random_state=42)
grid_search = GridSearchCV(logic,parameters,cv=5,scoring='accuracy')
```

```python
grid_search.fit(X_train,y_train)
```

```python
print("Best Parameters: ",grid_search.best_params_)
best_model = grid_search.best_estimator_
```

```python
y_pred = best_model.predict(X_test)
y_proba = best_model.predict_proba(X_test)[:,-1]
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score,classification_report,confusion_matrix
```

```python
print("Accuracy of the model: " ,accuracy_score(y_test,y_pred))
```

```python
print("Precision: ",precision_score(y_test,y_pred))
```

```python
print("Recall : ",recall_score(y_test,y_pred))
```

```python
print("F1 score: ",f1_score(y_test,y_pred))
```

```python
print("ROC-AUC: ",roc_auc_score(y_test,y_pred))
```

```python
print("Classification Report: ",classification_report(y_test,y_pred))
```

```python
print("Confusion matrix: ",confusion_matrix(y_test,y_pred))
```
