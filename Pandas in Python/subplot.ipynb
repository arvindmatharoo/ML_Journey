```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [18,37,57,39,58]
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("day1")

x = [1,2,3,4,5]
y=[10,40,20,40,50]
plt.subplot(2,2,2)
plt.bar(x,y)
plt.title('day2')

x = [1,2,3,4,5]
y = [10,20,30,44,12]
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title("day3")

x = [1,2,3,4,5]
y = [30,44,55,66,11]
plt.subplot(2,2,4)
plt.scatter(x,y)
plt.title("day4")
plt.suptitle("grocery sale")

plt.show()
```

```python

import numpy as np

# Generate sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a 1x2 grid of subplots (1 row, 2 columns)
fig, ax = plt.subplots(1, 2, figsize=(12, 4))

# Plot on the first subplot
ax[0].plot(x, y1, color='blue', label='Sine')
ax[0].set_title('Sine Wave')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[0].legend()

# Plot on the second subplot
ax[1].plot(x, y2, color='green', label='Cosine')
ax[1].set_title('Cosine Wave')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the figure
plt.show()
```
