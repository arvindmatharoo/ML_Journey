```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig,ax = plt.subplots(2,2,figsize =(10,8))
x = np.linspace(0,10,100)
y = np.sin(x)
y2 = np.cos(x)
y3 = np.exp(-x)
y4 = x**2
ax[0,0].plot(x,y,color = "red")
ax[0,0].set_title("Sine wave")
ax[0,1].plot(x,y2,color= "orange")
ax[0,1].set_title("Cosine wave")
ax[1,0].plot(x,y3,color = "purple")
ax[1,0].set_title("Exponential decay")
ax[1,1].plot(x,y4,color = "pink")
ax[1,1].set_title("Quadratic function")

plt.tight_layout()
plt.show()
```
