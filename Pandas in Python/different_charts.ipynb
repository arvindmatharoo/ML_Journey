```python
import matplotlib.pyplot as plt
import numpy as np 
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,color = "cyan")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("sine wave")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
x = ['A','B','C','D']
y = [10,12,3,40]
plt.bar(x,y,color = "purple")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("BAR GRAPH")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y,color = "pink",marker = 's')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("scatter chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.random.rand(100)
plt.hist(data,color = "blue" , edgecolor = "black")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
data = [np.random.rand(100) for _ in range(4)]
plt.boxplot(data , labels = ['group1' , 'group2' , 'group3', 'group4'])
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("box chart")
plt.show()
```
